```python
#it's easy to see that the date(DDMMYYYY) is unneccary, so we'll remove it. we'll be left with the time format
def change_time(x):
  try:  #x is a string 
       a = x[11:-1]
       a = a.split(":")
       b =  int(a[0])*3600 + int(a[1])*60 + int(a[2])
       return b
  except:
    return np.nan
for i in ['order_time','allot_time','accept_time', 'pickup_time','delivered_time', 'cancelled_time']:
  train[i] = train[i].apply(lambda x: change_time(x ))
```

```python
#this is a MAASIVE bias in data.
#If directly inputted into our model, the said model will ALWAYS predict cancelled = 0. Keeping the problem statement in mind, we must balance the data 
train = pd.concat([train[train['cancelled'] > 0], train[train['cancelled'] == 0].sample(int(len(train[train['cancelled'] > 0])*1.5))], axis = 0).sample(frac = 1)
```

```python
#unfortunately we lost a lot of data, but we at least have a better representation 
#now we'll try delete/alter the features
del train['order_id'] #for obvious reasons
del train['order_date'] #in practical cases, the date won't matter. The deliveries will most definitely be delivered on the sae date
del train['rider_id'] #obvious yet again
#same thing for test set
del test['order_id']
del test['order_date']
del test['rider_id']
```

```python
def REMOVE_STRING(x): #I spent 5 hours on this part alone
  if isinstance(x, str) is True:
    return np.nan
  elif isinstance(x, int)|isinstance(x,float) is True :
    return float(x)
for i in ['order_time', 'allot_time', 'accept_time', 'pickup_time',
       'delivered_time', 'first_mile_distance', 'last_mile_distance',
       'alloted_orders', 'delivered_orders', 'cancelled', 'undelivered_orders',
       'lifetime_order_count', 'session_time', 'cancelled_time']: 

 train[i] = train[i].apply(lambda x: REMOVE_STRING(x))
```

```python
train.head()
```

```python
from sklearn.experimental import enable_iterative_imputer 
from sklearn.impute import IterativeImputer #unlike simple imputation, this one predicts the values based on other parameters of record.
imp = IterativeImputer(max_iter=10, random_state=0) 
values1 = train[['order_time', 'allot_time', 'accept_time', 'pickup_time',
       'first_mile_distance', 'last_mile_distance',
       'alloted_orders', 'delivered_orders', 'cancelled', 'undelivered_orders',
       'lifetime_order_count']].values
train[[ 'pickup_time',
       'delivered_time', 'session_time', 'cancelled_time']] =train[[ 'pickup_time',
       'delivered_time', 'session_time', 'cancelled_time']].fillna(0)
t_values = imp.fit_transform(values1)
train[ ['order_time', 'allot_time', 'accept_time', 'pickup_time',
       'first_mile_distance', 'last_mile_distance',
       'alloted_orders', 'delivered_orders', 'cancelled', 'undelivered_orders',
       'lifetime_order_count']] = t_values
train.head()
```

```python
train = train.assign( allot_time = train['allot_time']-train['order_time'], pickup_time = train['pickup_time']-train['order_time'], accept_time = train['accept_time']-train['order_time'], cancelled_time = train['cancelled_time']-train['order_time']  )
train['total_time'] = train['allot_time']+train['accept_time']
train.head()
```

```python
from sklearn.pipeline import Pipeline
from sklearn import model_selection, metrics, svm
from sklearn.preprocessing import StandardScaler
pipeline = Pipeline([
     ('scaler', StandardScaler()),
    ('svm', svm.SVC(kernel='rbf', C=1.2, gamma=0.2))
])
```

```python
plt.figure(figsize =(20,5))
train.boxplot()
```

```python
from sklearn import model_selection
Xt, Xv, yt, yv = model_selection.train_test_split(train[['order_time', 'allot_time', 'accept_time',
       'first_mile_distance', 'last_mile_distance',
       'alloted_orders', 'delivered_orders', 'undelivered_orders',
       'lifetime_order_count','total_time']], train['cancelled'], test_size = 0.2,random_state = 7)
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
pipeline.fit(Xt, yt)
predictions = pipeline.predict(Xv)
print(metrics.accuracy_score(yv, predictions))
print(confusion_matrix(yv, predictions))
print(classification_report(yv, predictions))
#that 99% from the screenshot came from a weird error, so I had to fix it. Unfortunately the precision drops down quite a bit
```
